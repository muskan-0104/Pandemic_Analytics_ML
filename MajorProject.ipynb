```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('covid_19_india.csv')
mask2=(data['State/UnionTerritory']=='Punjab')
data=data.loc[mask2]
#data['Date'] = pd.to_datetime(data.Date)

#data=data.sort_values(by='Date')

ax = plt.subplot(111)
ax.plot(data['Date'],data['Confirmed'],color='Yellow',linewidth=3)
ax.plot(data['Date'],data['Cured'],color='Green',linewidth=3)
ax.plot(data['Date'],data['Deaths'],color='Red',linewidth=3)
ax.set_xlim(0, 120)
plt.xlabel("Date")
plt.ylabel("Covid Cases")
plt.title("Rise in Cases(Date Sorted) in Punjab")
plt.grid(True)
#plt.savefig('Covid_Graph1.png')
plt.figure(figsize=(6,10))
ax.legend();
plt.show()

```
