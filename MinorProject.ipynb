```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('covid_19_india.csv')
data['Date'] = pd.to_datetime(data['Date'])
start_date = '2020-06-01'
end_date = '2020-06-30'
mask = (data['Date'] >= start_date) & (data['Date'] <= end_date)
data = data.loc[mask]
mask2=(data['State/UnionTerritory']=='Maharashtra')
data=data.loc[mask2]
data=data.sort_values('Date')
```

```python
ax = plt.subplot(111)
ax.plot(data['Date'].dt.day,data['Confirmed'],color='Yellow',linewidth=3)
ax.plot(data['Date'].dt.day,data['Cured'],color='Green',linewidth=3)
ax.plot(data['Date'].dt.day,data['Deaths'],color='Red',linewidth=3)
plt.xlabel("Date")
plt.ylabel("Covid Cases")
plt.title("Rise in Cases(Date Sorted) in Maharshtra")
plt.grid(True,color="grey")
plt.savefig('Covid_Graph1.png')

ax.legend();
plt.show()
```

```python
data_n=pd.read_csv('covid_19_india.csv')
data_n['Date'] = pd.to_datetime(data_n['Date'])
start_date_n = '2020-06-04'
end_date_n = '2020-06-30'
mask_n = (data_n['Date'] >= start_date_n) & (data_n['Date'] <= end_date_n)
data_n = data_n.loc[mask_n]
mask2_n=(data_n['State/UnionTerritory']=='Rajasthan')
data_n=data_n.loc[mask2_n]
data_n=data_n.sort_values('Date')
```

```python
ax1 = plt.subplot(111)
ax1.plot(data['Date'].dt.day,data['Confirmed'],color='Yellow',linewidth=3,label="Maharashtra")
ax1.plot(data_n['Date'].dt.day,data_n['Confirmed'],color='Green',linewidth=3,label="Rajasthan")
plt.xlabel("Date")
plt.ylabel("Confirmed Covid Cases")
plt.title("Confirmed Cases Rajathan vs Maharashtra")
plt.grid(True,color="grey")
#plt.savefig('Covid_Graph1.png')

ax1.legend();
plt.show()
```
