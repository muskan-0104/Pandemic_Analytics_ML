```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('covid_19_india.csv')
#data['Date']=pd.to_datetime(data['Date'])
data['Date'] = pd.to_datetime(data['Date'])
start_date = '2020-06-01'
end_date = '2020-06-30'
mask = (data['Date'] >= start_date) & (data['Date'] <= end_date)
data = data.loc[mask]
mask2=(data['State/UnionTerritory']=='Maharashtra')
data=data.loc[mask2]
data=data.sort_values('Date')
data
```

```python
ax = plt.subplot(111)
ax.plot(data['Date'].dt.day,data['Confirmed'],color='Yellow',linewidth=3)
ax.plot(data['Date'].dt.day,data['Cured'],color='Green',linewidth=3)
ax.plot(data['Date'].dt.day,data['Deaths'],color='Red',linewidth=3)
plt.xlabel("Date")
plt.ylabel("Covid Cases")
plt.title("Rise in Cases")
plt.grid(True,color="grey")
plt.savefig('Covid.png')

ax.legend();
plt.show()
```
