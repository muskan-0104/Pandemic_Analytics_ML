```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
matplotlib.style.use('ggplot')

x=pd.read_csv('covid_19_india.csv')
x['Date'] = pd.to_datetime(x['Date'])
y=pd.read_csv('us_covid19_daily.csv')
y['lastModified'] = pd.to_datetime(y['lastModified'])
xy = pd.merge(left=x, right=y, left_on='Date', right_on='lastModified')
xy
plt.scatter(xy['Cured'], xy['recovered'])
plt.show()
```
